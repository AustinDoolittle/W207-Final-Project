### Cleaning
It appears that there are NaN values sprinkled through the dataset. Based on the codebook, it looks like this is split into either not having a specific feature or just missing data. 

An example of the first case is the `airconditionigtypeid` which has a value `NaN` when there is not air conditioning in the house. We don't want to remove all instances of no air conditioning as that will effect our final classification. Instead, we will treat these as categorical variables and use dummy variables for each categorical value. These columns are assumed to be those in the codebook that have the "(if any)" qualifier in their description.

The second case just deals with missing values. In the below code, we can check if we have any instances of a complete row excluding the columns that us NaN as a category.

```python
# These are the columns where NaNs are considered a category and should not be removed
no_remove_nans = set(['airconditioningtypeid', 'decktypeid', 'fireplacecnt', 'poolcnt', 'regionidcity'])

# get the remaining columns
remove_nans = set(props.columns) - no_remove_nans

# check if there are any columns that have a complete row
(~props[remove_nans].isna().any('columns')).any()
```
